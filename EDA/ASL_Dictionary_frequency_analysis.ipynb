## Next Step : Now let's see the frequency of each character. Will be useful as we are looking at fingerspelling in ASL we have to know what we will feed our machine learning to learn from.

```python
df = pd.read_csv('d:/Bureau/ASLCapstone/train.csv/train.csv')
df
```

```python
# Dictionary to store counts of each digit
digit_counts = {}

# Loop through each digit from 0 to 9
for digit in range(10):
    # Convert the digit to a string and use 'str.count()' to count occurrences in the 'phrase' column
    count_of_digit = df['phrase'].str.count(str(digit)).sum()
    # Store the count in the dictionary with the digit as the key
    digit_counts[str(digit)] = count_of_digit

# Print the counts of each digit
for digit, count in digit_counts.items():
    print("Number of occurrences of '{}' in the 'phrase' column: {}".format(digit, count))
```

```python
# Dictionary to store counts of each letter
letter_counts = {}

# Loop through each letter of the alphabet (both uppercase and lowercase)
for letter in 'fghijklmnopqrstuvwxyz':
    # Use 'str.count()' to count occurrences of the letter in the 'phrase' column
    count_of_letter = df['phrase'].str.count(letter).sum()
    # Store the count in the dictionary with the letter as the key
    letter_counts[letter] = count_of_letter

# Print the counts of each letter
for letter, count in letter_counts.items():
    print("Number of occurrences of '{}' in the 'phrase' column: {}".format(letter, count))
```

```python
char_counts = {}

# List of characters to count occurrences
characters = ["~", "", "!", "#", "$", "%", "&", "'", "(", ")", "*", "+", ",", "-", ".", "/", ":", ";", "=", "?", "@", "[", "_"]

# Loop through each character
for char in characters:
    # Escape the special characters in the regular expression pattern
    escaped_char = re.escape(char)
    # Use 'str.count()' to count occurrences of the character in the 'phrase' column
    count_of_char = df['phrase'].str.count(escaped_char).sum()
    # Store the count in the dictionary with the character as the key
    char_counts[char] = count_of_char

# Print the counts of each character
for char, count in char_counts.items():
    print("Number of occurrences of '{}' in the 'phrase' column: {}".format(char, count))
```
